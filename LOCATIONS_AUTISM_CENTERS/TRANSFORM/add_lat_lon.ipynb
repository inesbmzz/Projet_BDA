```python
import pandas as pd
```

```python
autism_center = pd.read_csv('autism_center.csv')

usa_lat_lon = pd.read_csv('usa_location.csv')
```

```python
len(autism_center)
```

```python
merged_df = pd.merge(autism_center, usa_lat_lon, left_on='city', right_on='name', how='left')
```

```python
len(merged_df)
```

```python
merged_df
```

```python
merged_df['latitude'] = merged_df['latitude_x'].fillna(merged_df['latitude_y'])
merged_df['longitude'] = merged_df['longitude_x'].fillna(merged_df['longitude_y'])
```

```python
merged_df.drop(columns=['latitude_x','longitude_x','latitude_x','state','latitude_y','longitude_y','name_y'], inplace=True)
```

```python
merged_df = merged_df.rename(columns={'name_x': 'name'})
```

```python
desired_order = ['name', 'country', 'city', 'address', 'latitude', 'longitude', 'info', 'free_quote_link']
merged_df = merged_df.reindex(columns=desired_order)
```

```python
merged_df.info()
```

```python
merged_df.to_csv('clean_autism_center.csv', index=False)
```
